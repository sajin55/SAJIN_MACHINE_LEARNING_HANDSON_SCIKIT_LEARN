```python
from sklearn.datasets import make_moons
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
def regularization(cr,gam):
 x,y=make_moons(n_samples=100,noise=0.15,random_state=42)
 svm=Pipeline([('scaler',StandardScaler()),('svm',SVC(C=cr,kernel='rbf',gamma=gam,max_iter=10000))])
 svm.fit(x,y)
 c=x[y==1][:,0]
 d=x[y==1][:,1]
 plt.plot(c,d,'o')
 e=x[y==0][:,0]
 f=x[y==0][:,1]
 plt.plot(e,f,'o')
 plt.xlabel('X1')
 plt.ylabel('X2')
 plt.legend(['1','0'])
 plt.title('gamma={}C={}'.format(gam,cr))
 x1s=np.linspace(-2,3,100)
 x2s=np.linspace(-1,2,100)
 x1,x2=np.meshgrid(x1s,x2s)
 x_new=np.c_[x1.ravel(),x2.ravel()]
 op_new=svm.predict(x_new).reshape(x1.shape)
 plt.contourf(x1,x2,op_new,alpha=0.15)
 plt.show()
regularization(0.001,0.1)
regularization(1000,0.1)
regularization(0.001,5)
regularization(1000,5)
```

```python
                                        #DECISION TREE VISULIZATION 2
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
import graphviz
iris=load_iris()
x=iris.data[:,2:]
y=iris.target
tree=DecisionTreeClassifier(max_depth=3)
tree.fit(x,y)
dotdata=export_graphviz(tree,out_file=None,feature_names=iris.feature_names[2:],class_names=iris.target_names,rounded=True,filled=True)
graph=graphviz.Source(dotdata)
graph
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
iris=load_iris()
x=iris.data[:,2:]
y=iris.target
tree=DecisionTreeClassifier(max_depth=3)
tree.fit(x,y)
print('class probabilities=',tree.predict_proba([[5,1.5]]))
print('class=',tree.predict([[1,0.9]]))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
import graphviz
from sklearn.datasets import make_moons
x,y=make_moons(n_samples=100,noise=0.25,random_state=53)
tree=DecisionTreeClassifier(random_state=42)
tree.fit(x,y)
tree1=DecisionTreeClassifier(random_state=42,min_samples_leaf=4)
tree1.fit(x,y)
dotdata=export_graphviz(tree,out_file=None,feature_names=['x1','x2'],class_names=['0','1'],rounded=True,filled=True)
graph=graphviz.Source(dotdata)
#graph.render('moondataset')
print('class probabilities(tree)=',tree.predict_proba([[5,1.5]]))
print('class(tree)=',tree.predict([[1,-0.9]]))
print('class probabilities(tree1)=',tree1.predict_proba([[5,1.5]]))
print('class(tree1)=',tree1.predict([[1,-0.9]]))
```
